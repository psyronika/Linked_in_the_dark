### BERTopic over time (2017, March - 2021, July)

```python
from bertopic import BERTopic
from sentence_transformers import SentenceTransformer
import pandas as pd
import pickle
import spacy
import numpy as np
import plotly.io as pio
pio.renderers.default='iframe'
#pio.renderers? -#make sure that this is set to 'iframe' so that figure displays
```

```python
#open data
with open('clean_df.pkl', 'rb') as f:
    df = pickle.load(f)
```

```python
#load model
topic_model = BERTopic.load("model_300K")
```

```python
#load topics
with open('topics.pkl', 'rb') as f:
    topics = pickle.load(f)
```

```python
docs = df.astype(str).texts.to_list()
timestamps = df.reset_index().date.to_list()
```

```python
## topics over time
#topics over time on the full dataset
#we employ global tuning and evolution tuning
topics_over_time = topic_model.topics_over_time(docs, topics, timestamps, global_tuning=True, evolution_tuning=True, nr_bins=20)
```

```python
#save topics over time to pickle
topics_over_time.to_pickle("topics_over_time.pkl")
```
